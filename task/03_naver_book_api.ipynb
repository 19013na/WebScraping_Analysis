##### 3-1. 네이버 책 검색 API 호출하기

```python
from dotenv import load_dotenv
import requests
import os
import json


def search_books(query): 
    # .env 파일 불러오기
    load_dotenv()

    client_id = os.getenv("NAVER_CLIENT_ID")
    client_secret = os.getenv("NAVER_CLIENT_SECRET")


    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret,
    }

    # query string 문자열을 dict 선언
    payload = {
        'query': query,
        'display': 50,
        'sort': 'sim'
    }

    url = 'https://openapi.naver.com/v1/search/book.json'
    
    # requests get(url, params, headers) 요청 
    res = requests.get(url, params=payload, headers=headers)

    # json 형식
    items_data = res.json()['items']
    return items_data
```

##### 질문 : 검색어로 찾은 책 목록을 JSON 파일로 저장하기

```python
def save_json(query):
    items_data = search_books(query)
# 질문 : 검색어로 찾은 책 목록을 JSON 파일로 저장하기
    first_file_name = 'data/shops.json'
    os.makedirs(os.path.dirname(first_file_name), exist_ok=True)

    # JSON 파일로 저장
    with open(first_file_name, 'w', encoding='utf-8') as f:
        json.dump(items_data, f, ensure_ascii=False, indent=4)
```

##### 질문 : books.json 파일을 Pandas DataFrame로 저장하기

```python
import pandas as pd

save_json("자바")
book_df = pd.read_json('data/books.json')
book_df.head(3)
```

##### 질문 :  검색어로  찾은  책 목록 출력하기

```python
import pandas as pd

def print_books(query):   
    items_data = search_books(query)
    return pd.DataFrame(items_data)


shop_df = print_books('파이썬')
shop_df.head()
```

#####  질문 :  검색어로  찾은  책 목록 중에서 가격이 2만원 이상인 책만 출력하기
* title,author,discount,publisher,pubdate 컬럼만 출력
* 가격은 descending (내림차순), index 초기화


```python
book_df = print_books('파랑')

book_df.loc[book_df['discount'].astype(int) >= 20000, book_df.columns.drop(['link', 'image', 'isbn', 'description'])]\
        .sort_values(by='discount', ascending=False).reset_index(drop=True)
```

##### 검색어로  찾은  책 목록 중에서 출판사가 인피니티북스인 책만 출력하기
* image , description 컬럼은 제외한 모든 컬럼 출력하기
* index 는 초기화


```python
book_df = print_books('자바')

book_df.loc[book_df['publisher'] == "인피니티북스", book_df.columns.drop(['image', 'description'])].reset_index(drop=True)
```
